```python
#REMOVE NOT NEEDED COLUMNS
#--------------------------
df1 = df.iloc[:,[0,1,3,4,5,12]]
```

```python
#CHECK UNIQUE VALUES OF COLUMN
#------------------------------
listvalcol = set(list(df.iloc[:,5]))
listvalcol

#NOTE ABSOLUTE VALUES ARE INCLUDED IN THE DATASET, THIS MEANS THAT RECORDS ARE REPEATED.
```

```python
#CREATE AN INDEX FOR ONLY 2020 and 2019 YEARS
#--------------------------------------------
regex_date = "2020.*|2019.*"
series_date = df1.iloc[:,0]
index_date = series_date.str.contains(regex_date)
index_date
```

```python
#REMOVE NOT NEEDED DATES
#--------------------------
df2 = df1[index_date]
```

```python
#CREATE AN INDEX FOR ONLY "to" VALUES
#--------------------------------------------
regex_Age = ".*to.*|65 years and over"
series_Age = df2.iloc[:,4]
index_Age = series_Age.str.contains(regex_Age)
index_Age
```

```python
#REMOVE NOT NEEDED Ages
#--------------------------
df3 = df2[index_Age]
```

```python
#CREATE AN INDEX TO EXCLUDE "CANADA" FROM LOCATIONS
#--------------------------------------------
regex_location = "[^Canada]"
series_location = df3.iloc[:,1]
index_location = series_location.str.contains(regex_location)
index_location
```

```python
#CREATE AN INDEX TO EXCLUDE "CANADA" FROM LOCATIONS
#--------------------------------------------
regex_location = "[^Canada]"
series_location = df3.iloc[:,1]
index_location = series_location.str.contains(regex_location)
index_location
```

```python
df4 = df3[index_location]
df4
```

```python
listvalcol = set(list(df4.iloc[:,4]))
listvalcol
```

```python
#CLEAN GENDER BOTH ONE LINE INDEXING
#-----------------------------------
index_clean_sex = df4.pipe(lambda x: x['Sex'] != "Both sexes")
df5 = df4[index_clean_sex]
df5
```

```python
#RENAME COLUMNS
#--------------
df5 = df5.rename({
                'REF_DATE': 'date',
                'GEO': 'location',
                'Type of work sought and search method': 'search_method',
                'Sex': 'gender',
                'Age group': 'age',
                'VALUE': 'count_value'
}, axis='columns')

df5
```

```python
#VALIDATION GROUP BY search_method
#--------------
df5.groupby(['date']).sum()
```

```python
#VALIDATION - PERCENTAGE OF TOTAL
#---------- -- -----
Total = df5['count_value'].sum()
x = df5.groupby(['location','gender','age']).sum()
x/Total
```

```python
#CHANGE count datatype to int currently is float64 because there are nan values
import numpy as np
df5['count_value'] = df5['count_value'].fillna(0).astype(np.int64)

#CHANGE date datatype date, currently is string
import datetime
df5['date'] = pd.to_datetime(df5['date'])
```

```python
#CHECK IF THERE ARE DUPLICATES IN THE ORIGINAL DATAFRAME
df.head(5)
```

```python
#CHECK IF THERE ARE DUPLICATED ROWS

def dupsAndclean(dfs):
    dups_var = dfs[dfs.duplicated(subset = None, keep = 'last')]
    clean_var = dfs.drop_duplicates()
    print("INFO")
    print(60*"-")
    print("Dataframe shape with no dups: ",clean_var.shape)
    print("Dataframe composed of duplicates shape: ",dups_var.shape)
    print("Total missing rows after removing dups ",dups_var.shape[0])
    
dupsAndclean(df1)
```

```python
#RETURN DUPLICATED ROWS AND CLEAN DATAFRAME
def removeDups(dfs):
    dups_var = dfs[dfs.duplicated(subset = None, keep = 'last')]
    clean_var = dfs.drop_duplicates()
    return [dups_var,clean_var]

#DUPLICATED ROwS
print(removeDups(df1)[0])

print(80*'-')

#CLEAN DATAFRAME
print(removeDups(df1)[1])
```
