```python
#https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=1410005801

import pandas as pd
df = pd.read_csv("14100058.csv")

df.shape
```

```python
df.head(5)
```

```python
#REMOVE NOT NEEDED COLUMNS
#--------------------------
df1 = df.iloc[:,[0,1,3,4,5,12]]

#CREATE AN INDEX FOR ONLY 2020 and 2019 YEARS
#--------------------------------------------
regex_date = "2020.*|2019.*"
series_date = df1.iloc[:,0]
index_date = series_date.str.contains(regex_date)
index_date

#REMOVE NOT NEEDED DATES
#--------------------------
df2 = df1[index_date]
```

```python
df2
```

```python
#CREATE AN INDEX FOR ONLY "to" VALUES
#--------------------------------------------
#regex_Age = ".*to.*24.*|.*to.*54.*|55 to 64 years|65 years and over|15 years and over"
regex_Age = "15 years and over"
series_Age = df2.iloc[:,4]
index_Age = series_Age.str.contains(regex_Age)
index_Age
```

```python
#REMOVE NOT NEEDED Ages
#--------------------------
df3 = df2[index_Age]
```

```python
#CHECK UNIQUE VALUES OF COLUMN
#------------------------------
listvalcol = set(list(df3.iloc[:,4]))
listvalcol
```

```python
#Ignore the red warning
df3['GEO'] = df3['GEO'].apply(lambda x: x.replace('Canada','ALL'))
df3['Sex'] = df3['Sex'].apply(lambda x: x.replace('Both sexes','ALL'))
df3['Age group'] = df3['Age group'].apply(lambda x: x.replace('15 years and over','ALL'))
```

```python
#CHECK UNIQUE VALUES OF COLUMN
#------------------------------
listvalcol = set(list(df3.iloc[:,1]))
listvalcol
```

```python
df3
```

```python
index_DATES_2019 = df3.pipe(lambda x: x['REF_DATE'] == "2020-07")
test = df3[index_DATES_2019]
index_GEO = test.pipe(lambda x: x['GEO'] == 'ALL')
test = test[index_GEO]
test.groupby(['REF_DATE','GEO','Age group','Sex']).sum()
```

```python
#RENAME COLUMNS
#--------------
df3 = df3.rename({
                'REF_DATE': 'date',
                'GEO': 'location',
                'Type of work sought and search method': 'search_method',
                'Sex': 'gender',
                'Age group': 'age',
                'VALUE': 'count_value'
}, axis='columns')

df3
```

```python
#VIEW FOR THE API'
df3.groupby(['date','location','age','gender']).sum()
```

```python
#RAW DATA ON TABULAR FORMAT, WILL AGREGATE USING DJANGO
#OUTPUT DATAFRAME
(df3).to_csv('unemployment_dataSet.csv', sep = ',', index=False)
```

```python
index_gender_all = df3.pipe(lambda x: x['gender'] != "ALL")
test = df3[index_gender_all]
index_location_all = test.pipe(lambda x: x['location'] != "ALL")
test = test[index_location_all]
#index_age_all = test.pipe(lambda x: x['age'] != "ALL")
#test = test[index_age_all]
```

```python
test.groupby(['date','location','age','gender']).sum()
```

```python
#BUILD RATIO COLUMN

test['percent'] = (test['count_value'] / test['count_value'].sum()) * 100
test.groupby(['date','location','age','gender']).sum()
```

```python
test.groupby('date').sum()
```

```python
#CHANGE count datatype to int currently is float64 because there are nan values
import numpy as np
test['count_value'] = 1000*test['count_value'].fillna(0).astype(np.int64)

#CHANGE date datatype date, currently is string
import datetime
test['date'] = pd.to_datetime(test['date'])
```

```python
final = test.groupby(['date','location']).sum().reset_index()
```
